```python
class ArrayStack:

    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

    def push(self, item):
        self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]


def splitTokens(exprStr):
    tokens = []
    val = 0
    valProcessing = False
    for c in exprStr:
        if c == ' ':
            continue
        if c in '0123456789':
            val = val * 10 + int(c)
            valProcessing = True
        else:
            if valProcessing:
                tokens.append(val)
                val = 0
            valProcessing = False
            tokens.append(c)
    if valProcessing:
        tokens.append(val)

    return tokens


def infixToPostfix(tokenList):
    prec = {
        '*': 3,
        '/': 3,
        '+': 2,
        '-': 2,
        '(': 1,
    }

    opStack = ArrayStack()
    postfixList = []
    
    for token in tokenList:
        # 숫자일 때(이전에 문자일때와 같음, 그대로 푸쉬)
        if type(token) is int:
            postfixList.append(token)
        # 가로 시작 받았을 때는 그대로 푸쉬
        elif token == '(':
            opStack.push(token)
        # 가로 끝 받았을 때는 가로 시작 받을 때까지 팝
        elif token == ')':
            for i in range(opStack.size()):
                pop_ = opStack.pop()
                if pop_ == '(':
                    break
                else:
                    postfixList.append(pop_)
            
        # 연산자를 받았을 때
        else:
            if opStack.isEmpty():
                opStack.push(token)
            else:
                peek_ = opStack.peek()
                if prec[peek_] < prec[token]:
                    opStack.push(token)
                elif prec[peek_] >= prec[token]:
                    for i in range(opStack.size()):
                        pop_ = opStack.pop()
                        if pop_ == '(':
                            break
                        else:
                            postfixList.append(pop_)
                    opStack.push(token)

                    
    while not opStack.isEmpty():
        postfixList.append(opStack.pop())
    
    
    return postfixList


def postfixEval(tokenList):
    valStack = ArrayStack()
    
    for token in tokenList:
        if type(token) is int:
            valStack.push(token)
        elif token in ['*', '/', '+', '-']:
            num2 = valStack.pop()
            num1 = valStack.pop()
            if token == '*':
                valStack.push(num1 * num2)
            elif token == '/':
                valStack.push(num1 / num2)
            elif token == '+':
                valStack.push(num1 + num2)
            elif token == '-':
                valStack.push(num1 - num2)

    
    return valStack.pop()


def solution(expr):
    tokens = splitTokens(expr)
    postfix = infixToPostfix(tokens)
    val = postfixEval(postfix)
    return val
```
